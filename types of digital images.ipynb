```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("C:/Users/User/Downloads/Image processing/cat.jpg")
```

```python
#rgb
cv2.imshow("img",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#bgr
cv2.imshow("bgr",cv2.cvtColor(img,cv2.COLOR_RGB2BGR))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#grayscale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("gray",gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#binary
ret,thres=cv2.threshold(gray,150,255,0)
cv2.imshow("bn",thres)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#histogram of red
import matplotlib.pyplot as plt
hist=cv2.calcHist([img],[2],None,[256],[0,256])
plt.plot(hist,color='r')
plt.show()
```

```python

#histogram of green
import matplotlib.pyplot as plt
hist=cv2.calcHist([img],[1],None,[256],[0,256])
plt.plot(hist,color='g')
plt.show()
```

```python
#histogram of blue
import matplotlib.pyplot as plt
hist=cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist,color='b')
plt.show()
```

```python
#histogam
from matplotlib import pyplot as plt 
colors = ('b','g','r') 
for i,color in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist,color = color)
plt.show()
```

```python
#histogam
from matplotlib import pyplot as plt 
hist = cv2.calcHist([gray],[0],None,[256],[0,256])
plt.plot(hist,color = "b")
plt.show()
```

```python
# Python program to explain splitting of channels

# Importing cv2
import cv2

# Reading the image using imread() function
image = cv2.imread('C:/Users/User/Downloads/Image processing/img.jpg')


# Using cv2.split() to split channels of coloured image
b,g,r = cv2.split(image)

# Displaying Blue channel image
# Blue colour is highlighted the most
cv2.imshow("Model Blue Image", b)

# Displaying Green channel image
# Green colour is highlighted the most
cv2.imshow("Model Green Image", g)

# Displaying Red channel image
# Red colour is highlighted the most
cv2.imshow("Model Red Image", r)

# Waits for user to press any key
cv2.waitKey(0)
```
