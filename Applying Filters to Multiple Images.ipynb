```python
import cv2
import glob
import matplotlib.pyplot as plt
path="D:/sidharth/images/*.*"
images=[cv2.imread(images,1) for images in
glob.glob(path)]
for i in range(len(images)):
    blur=cv2.GaussianBlur(images[i],(5,5),0)
    average_blur = cv2.blur(images[i], (5, 5))
    median_blur = cv2.medianBlur(images[i], 5)
    plt.title("Image"+str(i+1)+" Gaussian")
    plt.imshow(blur)
    plt.show()
    plt.title("Image"+str(i+1)+" Average")
    plt.imshow( average_blur)
    plt.show()
    plt.title("Image"+str(i+1)+" Median")
    plt.imshow(median_blur )
    plt.show()
```
