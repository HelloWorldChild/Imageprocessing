```python
im=cv2.imread("C:/Users/User/Downloads/Image processing/dora.webp")
```

```python
import cv2
#finding the dimensions
(h,w)=im.shape[:2]
#finding center of the image
(x,y)=(w // 2,h // 2)
#obtain the matrix of the image cv2.getRotationMatrix2D((midpoint of image),angle,scale)
m=cv2.getRotationMatrix2D((x,y),45,1.0)
#rotate=cv2.warpAffine(imagename,matrix,coloumns and rows)
rotate=cv2.warpAffine(im,m,(w,h))
plt.imshow(rotate)
```
